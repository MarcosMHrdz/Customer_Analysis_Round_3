Muestre un gráfico del número total de respuestas.


```python
response_counts = df['Response'].value_counts()
response_counts
```

```python
# Datos
response_labels = ['No','Yes']
response_counts = [7826, 1308]
 
# Gráfico de barras
plt.bar(response_labels, response_counts, color=['blue', 'green'])
plt.title('Número total de respuestas')
plt.xlabel('Respuesta')
plt.ylabel('Cantidad')
plt.show()
```

Muestre un gráfico de la tasa de respuesta por canal de ventas.

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Sales Channel', hue='Response')
plt.title('Tasa de respuesta por canal de ventas')
plt.xlabel('Sales Channel')
plt.ylabel('Count')
plt.legend(title='Response', loc='upper right')
plt.show()
```

Muestre un gráfico de la tasa de respuesta por el monto total del reclamo.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Response', y='Total Claim Amount')
plt.title('Tasa de respuesta por monto total del reclamo')
plt.xlabel('Response')
plt.ylabel('Total Claim Amount')
plt.show()
```

Muestre una gráfica de la tasa de respuesta por ingreso.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Response', y='Income')
plt.title('Tasa de respuesta por ingreso')
plt.xlabel('Response')
plt.ylabel('Income')
plt.show()
```
